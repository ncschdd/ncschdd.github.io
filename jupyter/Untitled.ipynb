```python
import tkinter as tk

class MyCal(tk.Tk): 
    def __init__(self):
        tk.Tk.__init__(self)
        self.title('Mini-Tr')
        self.s_buttons=['1','2','3','+',
                        '4','5','6','-',
                        '7','8','9','*',
                        '.','0','pi','/',
                        '(',')','<','=']
        self.init_label()
        self.init_buttons()
        self.update_label
        
    def init_buttons(self):
        self.b_buttons=[]
        for s in self.s_buttons:
            self.b_buttons.append(
            tk.Button(self,
                      text=s))
            self.b_buttons[-1].bind("<Button-1>",self.button_pressed)
        r,c=1,0
        
        for b in self.b_buttons:
         #   b.bind("<Button-1>",
         #          self.button_pressed)
            b.grid(row=r,column=c)            
            if c<3:
                c+=1
            else:
                r+=1
                c=0

    def button_pressed(self, event):
        self.update_label(
            event.widget['text'])

    def init_label(self):
        self.buttoninhalt = tk.Label(self,text='')
        self.buttoninhalt.grid(row=0,column=0,columnspan=4)
        
    def update_label(self, char, sym=None):
        if char == '=':
            a = self.buttoninhalt['text']
            self.buttoninhalt['text'] = str(eval(a))
        elif char == '<':
            self.buttoninhalt['text'] = ''
        else:
            self.buttoninhalt['text'] += char

app = MyCal()
app.mainloop()
```
